```python
import json
from deep_translator import GoogleTranslator
import time

# JSON 파일 경로 설정
input_path = 'univ-domains.json'
output_path = 'univ-domains-ko.json'

# JSON 파일 읽기
with open(input_path, 'r', encoding='utf-8') as f:
    data = json.load(f)

# 각 항목의 'name' 필드를 번역
translator = GoogleTranslator(source='en', target='ko')

for entry in data:
    original_name = entry['name']
    try:
        translated_text = translator.translate(original_name)
        entry['name'] = translated_text
        print(f"{original_name} → {translated_text}")
        time.sleep(0.5)  # 과도한 요청 방지
    except Exception as e:
        print(f"'{original_name}' 번역 중 오류 발생: {e}")
        entry['name'] = original_name  # 오류 발생 시 원래 이름 유지

# 번역된 데이터를 새로운 JSON 파일로 저장
with open(output_path, 'w', encoding='utf-8') as f:
    json.dump(data, f, ensure_ascii=False, indent=2)

print(f"'{output_path}'에 번역된 JSON 파일이 저장되었습니다.")
```
